### Example.1
##### Write a function calculate_area(radius) that calculates the area of a circle given its radius (Area=πr2). Use π=3.14159. 

***Programing Code***

```python
def calculate_area(radius): #defining the function
    pi = 3.14159# defining the value of pi
    area = pi * (radius ** 2)# area of the circle
    return area# defining the function to return the area

radius = float(input("Enter the radius of the circle: "))# taking the radius as input from the user
area = calculate_area(radius)# calling the function
print(f"The area of the circle with radius {radius} is: {area}")# printing the area of the circle
```

### Example 1.2
##### Create a class Rectangle with attributes width and height. Add a method area() that calculates the area of the rectangle. 

***Programing Code***

```python
class Rectangle:# defining the class
    def __init__(self, width, height):# defining the constructor
        self.width = width# defining the width of the rectangle
        self.height = height# defining the height of the rectangle

    def area(self):# defining the method to calculate the area of the rectangle
        return self.width * self.height# returning the area
    
rectangle = Rectangle(4, 10)# creating an object of the class
area = rectangle.area()# calling the method
print(f"The area of the rectangle with width {rectangle.width} and height {rectangle.height} is: {area}")# printing the area of the rectangle
    
```

#### Example 1.4
##### Write a program that reads a list of numbers from a file "numbers.txt" (one number per line) and calculates their sum. 

***Programing Code***

```python
# Step 1: Create the file and write numbers to it
with open('numbers.txt', 'w') as file:
    file.write('1\n2\n3\n4\n5\n')

# Step 2: Read the file and sum the numbers
total = 0
with open('numbers.txt', 'r') as file:
    numbers = file.readlines()
    for number in numbers:
        total += int(number.strip())

# Step 3: Print the result
print('The sum of the numbers in the file is:', total)
```

#### Example 1.5 
##### Write a program that asks the user to enter a number and calculates its square root. Handle the ValueError if the user enters a negative number. 

***Programing Code***

```python
import math

try:
    # Ask the user for input
    number = float(input('Enter a number: '))

    # Raise error if number is negative
    if number < 0:
        raise ValueError('Cannot calculate square root of a negative number.')

    # Calculate and display square root
    print(f'The square root of {number} is: {math.sqrt(number)}')

except ValueError as e:
    # Catch and print any ValueError
    print(e)

finally:
    # Always executed
    print('Execution completed.')
```
