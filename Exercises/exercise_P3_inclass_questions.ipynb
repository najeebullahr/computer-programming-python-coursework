### In Class_Exercise1

#### Create	a	Series from	a	dictionary.
##### Create three different pandas Series
###### one from a list of your favorite colors,
###### one from a NumPy array of random numbers with a labeled index,
###### one from a dictionary mapping days of the week to the number of hours you slept.
###### Print each Series.

### Programing Code

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
# 1. Series from a List of Favorite Colors
# Creating a Series from a simple Python list
colors = ['Red', 'Blue', 'Green', 'Black', 'White']

# Creating the Series
series_colors = pd.Series(colors)

# Printing the Series
print("Series from a List of Favorite Colors:")
print(series_colors)
print("\n")  # Print a new line for separation

# 2. Series from a NumPy array of random numbers with a labelled index
# Creating a NumPy array of random numbers
random_numbers = np.random.rand(5)

# Defining custom labels for the Series index
labels = ['a', 'b', 'c', 'd', 'e']

# Creating the Series with the random numbers and custom index
series_random = pd.Series(random_numbers, index=labels)

# Printing the Series
print("Series from a NumPy Array with Labelled Index:")
print(series_random)
print("\n")

# 3. Series from a dictionary mapping days of the week to hours slept
# Creating a dictionary for days and hours slept
sleep_hours = {
    'Monday': 3,
    'Tuesday': 9,
    'Wednesday': 6,
    'Thursday': 5,
    'Friday': 7,
    'Saturday': 11,
    'Sunday': 12
}

# Creating the Series from dictionary
series_sleep = pd.Series(sleep_hours)

# Printing the Series
print("Series from a Dictionary of Sleep Hours:")
print(series_sleep)
```

### In_ClassExample3
#### Construct a	DataFrame from	a	NumPy	array.
###### Take the DataFrame you created in the in-class exercise 2 and reindex it to include another student.
###### Fill any missing values with 'N/A’.Also, reindex the columns to include a 'Grade' column, filling missing values with 0.
###### Print the modified DataFrame.

### Programing Code


```python
import pandas as pd
import numpy as np

# Step 1: Create the original DataFrame (same as previous exercise)
student_data = np.array([
    ['Ali Hassan', 'Computer Science', 3.9],
    ['Fatima Zahra', 'Biology', 3.6],
    ['Mohammed Salah', 'Engineering', 3.5],
    ['Sara Ahmed', 'Economics', 3.7],
    ['Huda Noor', 'Chemistry', 3.2],
    ['David Brown', 'History', 3.4],
    ['Isabella Lewis', 'Psychology', 3.8],
    ['Ahmed Mansoor', 'Mathematics', 3.3],
    ['Aisha Kareem', 'Physics', 3.9],
    ['Jack Robinson', 'Political Science', 3.6]
])

# Create DataFrame
students_df = pd.DataFrame(student_data, columns=['Name', 'Major', 'GPA'])

print("Original Students DataFrame:")
print(students_df)
print("\n")

# Step 2: Reindex to include another student (add new index)
# Generate new index by adding one more (current index is 0 to 9, so next is 10)
new_index = list(students_df.index) + [10]

# Reindex the DataFrame and fill missing values with 'N/A'
students_df = students_df.reindex(new_index, fill_value='N/A')

print("After adding new student row with N/A values:")
print(students_df)
print("\n")

# Step 3: Reindex columns to include a new 'Grade' column
# Add 'Grade' column with fill value 0
students_df = students_df.reindex(columns=['Name', 'Major', 'GPA', 'Grade'], fill_value=0)

print("Final Modified DataFrame with Grade column and missing values filled:")
print(students_df)
```

### In_ClassExample4
#### Construct	a	DataFrame from	a	NumPy	array.
##### Using	the	DataFrame	from	the	in-class	exercise	3:
###### Drop	the	row you	added	in	the	reindexing	step.	
###### Then,	drop	the	'Grade'	column	you	added.
###### Print	the	DataFrame	after	each	drop	operation.

### Programing Code

```python
import pandas as pd
import numpy as np

# ------------------------------------------
# Step 1: Re-create the DataFrame from previous exercise (with extra row and Grade column)

student_data = np.array([
    ['Ali Hassan', 'Computer Science', 3.9],
    ['Fatima Zahra', 'Biology', 3.6],
    ['Mohammed Salah', 'Engineering', 3.5],
    ['Sara Ahmed', 'Economics', 3.7],
    ['Huda Noor', 'Chemistry', 3.2],
    ['David Brown', 'History', 3.4],
    ['Isabella Lewis', 'Psychology', 3.8],
    ['Ahmed Mansoor', 'Mathematics', 3.3],
    ['Aisha Kareem', 'Physics', 3.9],
    ['Jack Robinson', 'Political Science', 3.6]
])

# Create initial DataFrame
students_df = pd.DataFrame(student_data, columns=['Name', 'Major', 'GPA'])

# Reindex to add new row
new_index = list(students_df.index) + [10]
students_df = students_df.reindex(new_index, fill_value='N/A')

# Add new 'Grade' column
students_df = students_df.reindex(columns=['Name', 'Major', 'GPA', 'Grade'], fill_value=0)

print("Original DataFrame with extra row and Grade column:")
print(students_df)
print("\n")

# Step 2: Drop the extra row (index 10)
students_df_dropped_row = students_df.drop(index=10)

print("DataFrame after dropping the extra row:")
print(students_df_dropped_row)
print("\n")

# Step 3: Drop the 'Grade' column
students_df_dropped_column = students_df_dropped_row.drop(columns=['Grade'])

print("Final DataFrame after dropping 'Grade' column:")
print(students_df_dropped_column)
```
