### In Class Practice1

#### we	are	to	create	a	pandas	DataFrame	to	Handle	Missing	Data

### Programing Code

```python
import pandas as pd
import numpy as np
# Modified data with missing values (np.nan)
data = {
    'C1': [10, np.nan, 30, np.nan, 50],
    'C2': [np.nan, 20, np.nan, 40, np.nan],
    'C3': [60, 70, 80, 90, 100]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

#### 1. Remove all rows containing any missing values

```python
df_no_missing_any = df.dropna()
print("\nRows without any missing values:\n", df_no_missing_any)
```

#### 2. Remove only the rows where all values are missing

```python
df_not_all_missing = df.dropna(how='all')
print("\nRows where not all values are missing:\n", df_not_all_missing)
```

####  3. Keep only the rows that have at least 2 non-missing values

```python
df_at_least_two = df[df.notna().sum(axis=1) >= 2]
print("\nRows with at least 2 non-missing values:\n", df_at_least_two)
```

#### In-Class Pratice2
#### Filling	Missing	Data
#### Using	the	DataFrame	from	In-Class	Practice	1:

### Programing Code

```python
import pandas as pd
import numpy as np
# Create DataFrame with missing values
data = {
    'C1': [1, np.nan, 3, np.nan, 5],
    'C2': [np.nan, 2, np.nan, 4, np.nan],
    'C3': [6, 7, 8, 9, 10]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

#### 1. Fill all missing values with the mean of each column

```python
df_mean_filled = df.fillna(df.mean(numeric_only=True))
print("\nFilled with column means:\n", df_mean_filled)
```

#### 2. Forward fill the missing values in col1 (C1)

```python
df_ffill_c1 = df.copy()
df['C1'] = df['C1'].ffill()
print("\nForward fill in 'C1':\n", df_ffill_c1)
```

#### 3. Backward fill the missing values in col2 (C2), limit to one consecutive missing value

```python
df_bfill_c2 = df.copy()
df['C2'] = df['C2'].bfill(limit=1)
print("\nBackward fill in 'C2' (limit=1):\n", df_bfill_c2)
```

### Programing Code

```python
import pandas as pd

data = pd.DataFrame({
    'C1': ['A', 'B', 'A', 'C', 'B', 'A'],
    'C2': [1, 2, 1, 3, 2, 4]
})

print("Original DataFrame:\n", data)
```

```python
import pandas as pd

# Step 1: Create a DataFrame from the dictionary
data_dict = {'C1': ['A', 'B', 'A', 'C', 'B', 'A'], 'C2': [1, 2, 1, 3, 2, 4]}
data = pd.DataFrame(data_dict)

# Step 2: Identify duplicate rows
duplicates = data.duplicated()
print("\nDuplicate rows (entire row match):\n", duplicates)

# Step 3: Remove all duplicate rows, keeping the first occurrence
no_duplicates = data.drop_duplicates()
print("\nDataFrame with duplicates removed (keep first):\n", no_duplicates)

# Step 4: Remove duplicates based on 'C1' only, keeping the last occurrence
no_c1_duplicates = data.drop_duplicates(subset='C1', keep='last')
print("\nDataFrame with 'C1' duplicates removed (keep last):\n", no_c1_duplicates)
```
