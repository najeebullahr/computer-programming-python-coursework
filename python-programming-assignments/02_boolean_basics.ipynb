Example Exercise 4

```python
money = int(input('Please Enter the Principal Amount?'))
interest = float(input('Please Enter the annual Interest Rate (%)?'))
y1 = money * (1 + interest/100)
y2 = money * ((1 + interest/100) ** 2)
y3 = money * ((1 + interest/100) ** 3)
print('Principal and Interest in the First Year:', y1)
print('Principal and Interest in the Second Year:', y2)
print('Principal and Interest in the Third Year:', y3)
```
