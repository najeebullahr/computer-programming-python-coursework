##### Please write	a	program	to	determine	whether	the	input	value	is	odd	or	even.	Usually,	the	remainder	is	obtained	by	dividing	by	2.	If	the	remainder	is	0,	the	input	value	is	even,	otherwise,	the	input	value	is	odd.

```python
# Ask the user to enter an integer number
num = int(input('Please Enter an Integer: '))

# Check if the number is odd or even using modulus operator
if num % 2:
    # If remainder is 1, the number is odd
    print(num, 'is an Odd Number.')
else:
    # If remainder is 0, the number is even
    print(num, 'is an Even Number.')
```

##### Triangle	Judgment
#### Problem	Description
###### Design	a	program	that	allows	you	to	enter	the	lengths	of	the	three	sides	of	a	triangle,	a,	b,	and	c,	respectively,	and	determine	if	the	triangle	is	a	triangle	based	on	the	fact	that	the	sum	of	any	two	sides	of	the	triangle	is	greater	than	the	third	side.

```python
# Input the lengths of the three sides of the triangle
a = int(input('Enter the Length A of the Triangle: '))
b = int(input('Enter the Length B of the Triangle: '))
c = int(input('Enter the Length C of the Triangle: '))

# Check the Triangle Inequality Theorem
if (a < b + c) and (b < a + c) and (c < a + b):
    # If condition is true, these sides can form a triangle
    print('Can be Formed into a Triangle.')
else:
    # If condition is false, they cannot form a triangle
    print('Can’t be Formed into a Triangle.')
```

### Example 5
#### Postage Calculation
##### Problem Description
 ###### A courier company uses weight as the basis for calculating postage, and the weight and postage are calculated in the following table.
 ###### Please write a program to help the courier to calculate the postage. The courier only needs to enter the weight and the program will calculate the postage automatically.

```python
# Input the weight of the item
w = float(input("Enter the Item’s Weight in kg: "))

# Check the weight range and print the corresponding postage cost
if w <= 10:
    print("Postage Required is 50 Dollars.")
elif w > 10 and w <= 15:
    print("Postage Required is 70 Dollars.")
elif w > 15 and w <= 20:
    print("Postage Required is 90 Dollars.")
elif w > 20 and w <= 25:
    print("Postage Required is 110 Dollars.")
else:
    print("Over 25 kg, Sorry! Can’t be shipped.")
```

### Example 7
#### BMI	Calculator
 ##### Problem	Description
 ###### MI	is	often	used	to	determine	the	degree	of	obesity,	BMI	is	equal	to	the	weight	(KG)	divided	by	the	square	of	the	height	(M),	there	will	be	a	classification	standard.	Assume	that	BMI	and	obesity	classification	are	as	follows.	
###### Please	write	a program	for	users	to	input	their	height	and	weight	to	display	the	BMI	value	and	obesity	level.

```python
# Input weight in kilograms
w = float(input('Enter your weight (KG): '))

# Input height in meters
h = float(input('Enter your height (M): '))

# Calculate BMI
bmi = w / (h * h)

# Display calculated BMI
print('BMI is', round(bmi, 2))  # rounded for better readability

# Classify BMI based on standard ranges
if bmi < 18:
    print('Underweight')
elif bmi < 24:
    print('Normal Weight')
elif bmi < 27:
    print('Overweight')
else:
    print('Obese')
```
