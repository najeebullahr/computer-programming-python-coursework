```python
x = "Hello World"  # String
x = 20  # Integer
x = 20.5  # Float
x = 7j  # Complex number
x = ["apple", "banana", "cherry"]  # List
x = ("apple", "banana", "cherry")  # Tuple
x = range(6)  # Range
x = {"name": "john", "age": 36}  # Dictionary (Fixed Syntax)
x = {"apple", "banana", "cherry"}  # Set
x = frozenset({"apple", "banana", "cherry"})  # Frozenset
x = True  # Boolean
x = b"Hello"  # Bytes
x = bytearray(5)  # Bytearray
x = memoryview(bytes(5))  # Memoryview
x = None  # NoneType
```

Sets {}

```python
animals = {'cat', 'dog'}

print('cat' in animals)  # Check if 'cat' is in the set
print('fish' in animals)  # Check if 'fish' is in the set

animals.add('fish')  # Add 'fish' to the set
print('fish' in animals)  # Check again

print(len(animals))  # Print the length of the set

animals.add('cat')  # Adding 'cat' again (no duplicates in sets)
print(len(animals))  # The length remains the same

animals.remove('cat')  # Remove 'cat' from the set
print(len(animals))  # Print the new length of the set
```
