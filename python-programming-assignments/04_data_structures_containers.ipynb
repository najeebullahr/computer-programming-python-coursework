# List Function (list)
#### Use “[ start: end: interval]” to cut the string from “start” to “end” (without the end character) by taking one character for each "interval" character.

##### 1. list[:] means to take each element of the list, if no end element is specified, the last element is used to end by default, if no start element is specified, the first element is used to start by default.

***programing Code***

```python
a = list('abcdefghijk')
print(a[:])
```

##### 2. list[start:] means to take all the elements from the list[start] to the end of the list. If no end element is specified, the last element ofthe list is used by default, including the last element.

##### 3. list[:end] means to take all the elements from the first element of the list to the first element of the specified element of the list[end], if no start element is specified, the first element of the list is used by default.****

***Programing Code***

```python
a = list('abcdefghijk') #list
print(a[:5])  #take the letter before 5
print(a[5:])  #take the letter after 5
print(a[:-2]) #take the letter before 2
print(a[-2:]) #take the letter after 2
```

##### 4. list[ start: end] means to take all the elements from the list[ start] element to the first element of the specified element of the list[ end].

***Programing Code***

```python
a = list('abcdefghijk')
print(a[0:4])
print(a[-5:-3])
```

##### 5. list[ start: end: interval] means to take all the elements of the list [ start] to the first element of the specified element of the list [ end], one element for every “interval”.

***Programing Code***

```python
a = list('abcdefghijk')
print(a[1:10:3])
print(a[-1:-4:-1])
```

##### 6. list[::-1] denotes the inverted series list, where the inverted series is the one in which the first element of the series is swapped with the last element. 

***Programing Code***

```python
a = list('abcdefghijk')#list
print(a[::-1])#reverse
```

##### 7. Use [:] with the function copy to copy the list. It will copy the list as two different objects that occupy different memory space than the original list.

***Programing Code***

```python
list1 = [1, 2, 3, 4]#list
list3 = list1[:]#copy
list3[2] = 19#modifying
print('list1=', list1)#Output: list1= [1, 2, 19, 4]
print('list3=', list3)#Output: list3= [1, 2, 19, 4]
list4 = list1.copy()#copy
list4[2] = 19#modifying
print('list1=', list1)#Output: list1= [1, 2, 19, 4]
print('list4=', list4)#Output: list4= [1, 2, 19, 4]
```

##### 8. Using the equal sign “=”, for example, "list1 = list2", means that list1 and list2 refer to the same object, and when the elements in list1 or list2 are modified, both list1 and list2 will change.

***Programing Code***

```python
 list1 = [1, 2, 3, 4]
 list2 = list1
 print('list1=', list1)
 print('list2=', list2)
 list1[2]=19
 print('list1=', list1)
 print('list2=', list2)
 list2[2]=18
 print('list1=', list1)
 print('list2=', list2)
```

 # Dictionary Function (dict)

##### When you use the "key" to search for the corresponding "value", you will find that the 
##### dictionary stores data in a different order than the order in which the data was entered 
##### when the dictionary was constructed.–For example: numbers, strings and tuples.
##### Dictionaries can add, delete, update and merge two dictionaries.

 ##### 1. Use “{}” to create a new dictionary, which represents an element as “key: value”.

***Programing Coce***

```python
# Creating an empty dictionary
dict1 = {}  # empty dictionary
print(dict1)  # Output: {}

# Creating a set with greetings in different languages
lang = {'Good Morning', '早安', 'Hello', '你好'}  # set of greetings
print(lang)  # Output: {'Good Morning', '早安', 'Hello', '你好'}
```

##### 3. Use “function get” to read "value" corresponding to the "key".
###### If the "key" does not exist in the dictionary, then return None and the program will continue. 
###### Add a second parameter to the get function, 
###### If the "key" does not exist in the dictionary, then return the data entered in the second parameter.

***Programing Coce***

```python
# Creating a dictionary with English to Chinese greetings
lang = {'Good Morning': '早安', 'Hello': '你好'}  # dictionary

# Accessing value using key 'Hello'
print('English word for 「Hello」is', lang.get('Hello'))  
# Output: English word for 「Hello」is 你好

# Trying to access a non-existing key 'Hi'
print('English word for 「Hi」is', lang.get('Hi'))  
# Output: English word for 「Hi」is None

# Providing a default message when key is not found
print('English word for 「Hi」is', lang.get('Hi', 'Not in the Dictionary'))  
# Output: English word for 「Hi」is Not in the Dictionary
```

##### 4. Use “Dictionary[key]=value” to modify individual elements and add new elements. 
###### If the key exists in the dictionary, modify the value corresponding to the key;
###### if the key does not exist in the dictionary, add the key corresponding to the value.

***Programing Coce***

```python
# Initial dictionary with greetings
lang = {'Good Morning': '早安', 'Hello': '你好'}

# Adding a new entry: 'Hi' -> '你好'
lang['Hi'] = '你好'
print(lang)  # Output: {'Good Morning': '早安', 'Hello': '你好', 'Hi': '你好'}

# Adding another entry: 'Student' -> '學生'
lang['Student'] = '學生'
print(lang)  # Output: {'Good Morning': '早安', 'Hello': '你好', 'Hi': '你好', 'Student': '學生'}
```

 #### Merge Two Dictionaries
##### Use "function update" to merge two dictionaries into one dictionary 
##### For example, 
###### dict1.update(dict2), if there are duplicate "keys", the "keys" and "values" of dict2 will replace the "keys" and "values" of dict1.

***Programing Coce***

```python
# First example: merging lang2 into lang1
lang1 = {'Hello': '你好'}
lang2 = {'Student': '學生'}
lang1.update(lang2)
print(lang1)  
# Output: {'Hello': '你好', 'Student': '學生'}

# Second example: updating lang1 with a new key from lang2
lang1 = {'Good Morning': '早安', 'Hello': '你好'}
lang2 = {'Hi': '你好'}
lang1.update(lang2)
print(lang1)  
# Output: {'Good Morning': '早安', 'Hello': '你好', 'Hi': '你好'}
```

 #### Copy Dictionaries
##### dict2=dict1.copy() copies dict1 to dict2, dict1, and dict2 refer to different dictionary objects,

###### if you change the elements of dictionary dict1, it will not change dictionary dict2.
###### If you use "dict2=dict1", then dict1 and dict2 refer to the same dictionary object,
###### if you change the elements of dictionary dict1, dictionary dict2 will also be changed.

```python
# Example 1: Assignment (both variables point to the same dictionary)
lang1 = {'Good Morning': '早安', 'Hello': '你好'}
lang2 = lang1  # lang2 is just another name for lang1
lang2['Hi'] = '你好'  # This change affects both lang1 and lang2
print('lang1 is', lang1)
print('lang2 is', lang2)

# Example 2: Copy (creates a separate copy of the dictionary)
lang1 = {'Good Morning': '早安', 'Hello': '你好'}
lang3 = lang1.copy()  # lang3 is a new, independent dictionary
lang3['Hi'] = '你好'  # This change only affects lang3
print('lang1 is', lang1)
print('lang3 is', lang3)
```

#### Use "for" to read 
##### Each Element of the Dictionary
###### Using "for" to read each element of the dictionary, with the "function items" of the dictionary, both "keys" and "values" are returned.
 
###### Dictionary's "function keys" will return the "key" and the dictionary's "function values" will return the "value".

***Programing Code***

```python
# Dictionary of greetings
lang = {'Good Morning': '早安', 'Hello': '你好'}

# Loop 1: Using .items() to get both key and value
for en, ch in lang.items():
    print('English is', en, 'Chinese is', ch)

# Loop 2: Using .values() to get all values
for ch in lang.values():
    print(ch)

# Loop 3: Using .keys() to get all keys and access their values
for en in lang.keys():
    print(en, lang[en])
```

 # Set

#### Stores the data without order
###### To find out whether the data exists, 
###### Elements in the set cannot be duplicated, 
###### Set will automatically delete the duplicated elements.

#### Adding and Modifying Collections
##### If you use "set()" or "{}" to create a new set
###### Set will automatically remove duplicate elements
##### "set()" can only use one parameter
###### which can be a string, tuples, a series, or a dictionary to create a collection

***Programing Code***

```python
# Creating a set from a list of integers
s = {1, 2, 3, 4}
print(s)  # Output: {1, 2, 3, 4}

# Creating a set from a tuple (mix of strings and integers)
s = set(('a', 1, 'b', 2))
print(s)  # Output: {'a', 1, 2, 'b'}

# Creating a set from a list with duplicates
s = set(['apple', 'banana', 'apple'])
print(s)  # Output: {'apple', 'banana'}

# Creating a set from dictionary keys
s = set({'Good Morning': '早安', 'Hello': '你好'})
print(s)  # Output: {'Good Morning', 'Hello'}

# Creating a set from a string (unique characters)
s = set('racecar')
print(s)  # Output: {'r', 'e', 'a', 'c'}
```

#### Add and Remove Collections
##### Use "add" to add collection elements and "remove" to delete collection elements.

```python
# A set from a string  gives me unique characters
s = set('tiger')
print(s)  # Output: {'g', 't', 'r', 'i', 'e'} (order may vary)

# Adding a new element 'z' to the set
s.add('z')
print(s)  # Output: {'g', 't', 'r', 'i', 'e', 'z'} (order may vary)

# Removing the element 't' from the set
s.remove('t')
print(s)  # Output: {'g', 'r', 'i', 'e', 'z'} (order may vary)
```
