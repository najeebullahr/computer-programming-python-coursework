```python
# Python if-else statement example

score = 80  # Assign a score

# Check if the score is 60 or above
if score >= 60:
    print("Pass")  # This will run if score is 60 or more
else:
    print("Fail")  # This will run if score is less than 60
```

### if	elif	else

```python
score = 70  # Given score

# Check score and assign grade
if score >= 90:
    print("A")        # Executes if score is 90 or above
elif score >= 60:
    print("Pass")     # Executes if score is between 60 and 89
else:
    print("Fail")     # Executes if score is below 60
```

### if	elif	else

```python
# Python if-elif-else to assign a letter grade

score = 90
grade = ""  # Initialize empty grade variable

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "E"

# Print the final grade
print(grade)
```

### Function	def
#### Prgraming Code

```python
# Function to calculate Future Value (FV)
# Formula: FV = PV * (1 + r) ^ n

def getfv(pv, r, n):
    fv = pv * ((1 + r) ** n)
    return fv

# Call the function with example values
fv = getfv(100, 0.1, 7)

# Print the result rounded to 2 decimal places
print(round(fv, 2))
```

### Programing Code

```python
# Future Value Calculation

amount = 100          # Initial amount in dollars
interest = 10         # Interest rate in percent
years = 7             # Number of years

# Convert percentage to decimal and calculate future value
# Formula: FV = amount * (1 + rate/100) ** years
future_value = amount * ((1 + (0.01 * interest)) ** years)

# Print the result rounded to 2 decimal places
print(round(future_value, 2))  # Output: 194.87
```

 ### Classes	and	Objects
#### class	MyClass:

```python
# Python class example with a method

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def myfunc(self):
        print("Hi, I am " + self.name)

# Create an instance of Person
p1 = Person("John", 30)

# Call the method and print attributes
p1.myfunc()           # Output: Hi, I am John
print(p1.name)        # Output: John
print(p1.age)         # Output: 30
```

### Programing Code

```python
class Vehicle:
    name = ""
    kind = "BMW Car"
    color = ""
    value = 100.00

    def description(self):
        desc_str = "%s is a %s %s worth $%.2f." % (self.name, self.color, self.kind, self.value)
        return desc_str

# Example usage
v1 = Vehicle()
v1.name = "X5"
v1.color = "Black"
print(v1.description())
```

### Programing Code

```python
# Define the Vehicle class first
class Vehicle:
    def __init__(self, name="", kind="car", color="", value=100.00):
        self.name = name
        self.kind = kind
        self.color = color
        self.value = value

    def description(self):
        return f"{self.name} is a {self.color} {self.kind} worth ${self.value:,.2f}."

# Create instances of Vehicle
car1 = Vehicle(name="BMW", kind="Convertible", color="Silver", value=50000.00)
car2 = Vehicle(name="Toyota", kind="Van", color="Red", value=25000.00)

# Print descriptions and names
print(car1.description())  # Output: BMW is a Silver Convertible worth $50,000.00.
print(car1.name)           # Output: BMW
print(car2.description())  # Output: Toyota is a Red Van worth $25,000.00.
print(car2.name)           # Output: Toyota
```
