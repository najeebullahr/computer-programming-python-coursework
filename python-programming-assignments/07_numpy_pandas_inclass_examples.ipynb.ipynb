```python
import pandas as pd
import numpy as np

# 1. Series from a List of Favorite Colors
colors = ['Blue', 'Green', 'Black', 'White', 'Red']
series_colors = pd.Series(colors)
print("Series from a List of Favorite Colors:")
print(series_colors)
print()

# 2. Series from a NumPy array of random numbers with a labelled index
random_numbers = np.random.rand(5)
labels = ['a', 'b', 'c', 'd', 'e']
series_random = pd.Series(random_numbers, index=labels)
print("Series from a NumPy Array with Labelled Index:")
print(series_random)
print()
# 3. Series from a dictionary mapping days of the week to hours slept
sleep_hours = {
    'Monday': 7,
    'Tuesday': 6,
    'Wednesday': 8,
    'Thursday': 5,
    'Friday': 6,
    'Saturday': 9,
    'Sunday': 8
}
series_sleep = pd.Series(sleep_hours)
print("Series from a Dictionary of Sleep Hours:")
print(series_sleep)
print()
```

#### In_Class Exercise

```python
import pandas as pd
import numpy as np

# Step 1: Construct a DataFrame from a NumPy array
student_data = np.array([
    ['Alice Johnson', 'Physics', 3.8],
    ['Benjamin Lee', 'Mathematics', 3.5],
    ['Clara Smith', 'Biology', 3.7],
    ['David Brown', 'Computer Science', 3.9],
    ['Ella Davis', 'Chemistry', 3.4],
    ['Frank Wilson', 'History', 3.2],
    ['Grace Miller', 'Economics', 3.6],
    ['Henry Clark', 'Philosophy', 3.1],
    ['Isabella Lewis', 'Political Science', 3.7],
    ['Jack Robinson', 'Engineering', 3.8]
])

# Creating the DataFrame
students_df = pd.DataFrame(student_data, columns=['Name', 'Major', 'GPA'])

# Step 2: Reindex the rows to include a new student
new_index = list(students_df.index) + [10]  # Original indices 0-9, add 10
students_df = students_df.reindex(new_index, fill_value='N/A')

# Step 3: Reindex the columns to include 'Grade' and fill missing values with 0
students_df = students_df.reindex(columns=['Name', 'Major', 'GPA', 'Grade'], fill_value=0)

# Step 4: Printing the modified DataFrame
print("Modified Students DataFrame with Reindexed Rows and Columns:")
print(students_df)
```

```python
import pandas as pd
import numpy as np

# Dropping from a Series
s = pd.Series(np.arange(5), index=['a', 'b', 'c', 'd', 'e'])
print("Original Series s:\n", s)

s_dropped = s.drop(['c', 'd'])
print("\nSeries after dropping 'c' and 'd':\n", s_dropped)

# Dropping from a DataFrame
df = pd.DataFrame(np.arange(16).reshape((4, 4)),
                  index=['Ohio', 'Colorado', 'Utah', 'New York'],
                  columns=['one', 'two', 'three', 'four'])
print("\nOriginal DataFrame df:\n", df)

df_dropped_rows = df.drop(['Colorado', 'Ohio'])
print("\nDataFrame after dropping rows 'Colorado' and 'Ohio':\n", df_dropped_rows)

# Dropping columns from a DataFrame
df_dropped_cols = df.drop('two', axis=1)
print("\nDataFrame after dropping column 'two':\n", df_dropped_cols)

df_dropped_multiple_cols = df.drop(['two', 'four'], axis='columns')
print("\nDataFrame after dropping columns 'two' and 'four':\n", df_dropped_multiple_cols)
```

```python
import pandas as pd
import numpy as np

s = pd.Series(np.arange(4), index=['a', 'b', 'c', 'd'])
print("Series s:\n", s)

print("\ns['b']:\n", s['b'])
print("\ns['a':'c']:\n", s['a':'c'])
print("\ns[['b', 'd']]:\n", s[['b', 'd']])
print("\ns[s > 1]:\n", s[s > 1])

df = pd.DataFrame(np.arange(16).reshape((4, 4)),
                  index=['Ohio', 'Colorado', 'Utah', 'New York'],
                  columns=['one', 'two', 'three', 'four'])
print("\nDataFrame df:\n", df)

print("\ndf['two']:\n", df['two'])
print("\ndf[['three', 'one']]:\n", df[['three', 'one']])

print("\ndf[2:4]:\n", df[2:4])

print("\ndf[df['three'] > 5]:\n", df[df['three'] > 5])

print("\ndf.loc['Colorado', ['two', 'three']]:\n", df.loc['Colorado', ['two', 'three']])

print("\ndf.iloc[2, [3, 0, 1]]:\n", df.iloc[2, [3, 0, 1]])
```

#### Example9

```python
import pandas as pd
import numpy as np

# Create the DataFrame
df = pd.DataFrame(np.random.randn(4, 3), columns=list('bcd'),
                  index=['Utah', 'Ohio', 'Texas', 'Oregon'])

# Define formatting function
f = lambda x: '%.2f' % x

# Element-wise function application
print("\nElement-wise function:\n", df.applymap(f))

# Define a simple function
def times_two(x):
    return x * 2

# Apply function to a Series
print("\nApplying to a Series:\n", df['b'].map(times_two))

# Apply function to columns (sum across rows)
print("\nApplying to columns (sum across rows):\n", df.apply(np.sum, axis=0))

# Apply function to rows (min across columns)
print("\nApplying to rows (min across columns):\n", df.apply(np.min, axis=1))
```

#### Example10

```python
import pandas as pd
import numpy as np

# Sorting a Series
x = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
print("Original Series x:\n", x)

print("\nSorted Series by index:\n", x.sort_index())
print("\nSorted Series by values:\n", x.sort_values())

# Sorting a DataFrame
df = pd.DataFrame(np.random.randn(4, 3),
                  index=['three', 'one', 'two', 'four'],
                  columns=['b', 'a', 'c'])
print("\nOriginal DataFrame df:\n", df)

print("\nSorted DataFrame by index:\n", df.sort_index())
print("\nSorted DataFrame by column 'a':\n", df.sort_values(by='a'))

# Ranking
x2 = pd.Series([7.5, -2.5, 7.5, 4.2, 0.4])
print("\nOriginal Series for Ranking x2:\n", x2)

print("\nRanked Series (default method='average'):\n", x2.rank())
print("\nRanked Series (method='first'):\n", x2.rank(method='first'))
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame with some missing values (np.nan)
df = pd.DataFrame([[1.4, np.nan], [7.1, -4.5],
                   [np.nan, np.nan], [0.75, -1.3]],
                  index=['a', 'b', 'c', 'd'],
                  columns=['one', 'two'])

print("\nOriginal DataFrame:\n", df)

# Sum of columns (ignores NaN by default)
print("\nSum of columns:\n", df.sum())

# Mean of columns (ignores NaN by default)
print("\nMean of columns:\n", df.mean())

# Descriptive statistics summary
print("\nDescriptive statistics:\n", df.describe())
```

#### Example13

```python
import pandas as pd
import numpy as np

# Create a DataFrame
df = pd.DataFrame([[1.4, -0.3], [7.1, -0.5],
                   [np.nan, np.nan], [0.75, -1.3]],
                  columns=['one', 'two'])

print("\nOriginal DataFrame:\n", df)

# Correlation matrix (Pearson correlation by default)
print("\nCorrelation matrix:\n", df.corr())

# Covariance matrix
print("\nCovariance matrix:\n", df.cov())

# Correlation between specific columns
print("\nCorrelation between 'one' and 'two':\n", df['one'].corr(df['two']))
```
