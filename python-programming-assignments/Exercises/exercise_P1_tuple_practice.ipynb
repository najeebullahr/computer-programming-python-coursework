Example Using list function.

```python
#Creating a list 
my_list = [1, "banana", 2.71, False] 
print(my_list)  # Output: [1, 'banana', 2.71, False] 

# Modifying elements 
my_list[1] = "apple" 
print(my_list)  # Output: [1, 'apple', 2.71, False] 

# Adding elements 
my_list.append("orange") 
print(my_list)  # Output: [1, 'apple', 2.71, False, 'orange']

#Removing elements 
my_list.remove(2.71) 
print(my_list)  # Output: [1, 'apple', False, 'orange'] 
```

Example Using of dict function.

```python
#Creating a dictionary 
my_dict = {"name": "Alice", "age": 25, "city": "New York"} 
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

# Accessing values 
print(my_dict["name"])  # Output: Alice 

# Modifying values 
my_dict["age"] = 26 
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'} 

# Adding new key-value pairs 
my_dict["email"] = "alice@example.com" 
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York', 'email': 'alice@example.com'} 
# Removing key-value 

del my_dict["city"] 
print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'email': 'alice@example.com'} 
```

Example Using the set Function

```python
# Creating a set 
my_set = {1, 2, 3, 4, 4, 5} 
print(my_set)  # Output: {1, 2, 3, 4, 5} (duplicates are removed) 
# Adding elements 
my_set.add(6) 
print(my_set)  # Output: {1, 2, 3, 4, 5, 6} 

# Removing elements 
my_set.remove(3) 
print(my_set)  # Output: {1, 2, 4, 5, 6}

# Set operations 
set1 = {1, 2, 3} 
set2 = {3, 4, 5}
print(set1.union(set2))  # Output: {1, 2, 3, 4, 5} 
print(set1.intersection(set2))  # Output: {3}
```

Exercise 3: 
1. Create two sets containing your favorite fruits and your friend's favorite fruits. 
2. Find the common fruits between the two sets. 
3. Combine the two sets and print the result.

```python
my_favorite_fruits={"apple","banana","cherry","watermelon","orange"} #set
my_friend_favorite_fruits={"apple","Mango","cherry","orange"}#set
print(my_favorite_fruits.intersection(my_friend_favorite_fruits))#output is {"apple","cherry","orange"}.
print(my_friend_favorite_fruits.union(my_favorite_fruits))#output is {"apple","banana","cherry","watermelon","orange","Mango"}.
```

```python
# Tuple Challenge 
numbers = (10, 20, 30, 40, 50) 
total = sum(numbers) 
print("Sum of tuple elements:", total)  # Output: Sum of tuple elements: 150 

# List Challenge 
import random 
random_numbers = [random.randint(1, 100) for _ in range(10)] 
random_numbers.sort() 
print("Sorted list:", random_numbers)  # Output: Sorted list: [sorted numbers] 

# Dictionary Challenge 
people = {"Alice": 25, "Bob": 30, "Charlie": 20, "David": 35, "Eve": 40} 
oldest_person = max(people, key=people.get) 
print("Oldest person:", oldest_person)  # Output: Oldest person: Eve 

# Set Challenge 
set1 = {1, 2, 3, 4, 5} 
set2 = {4, 5, 6, 7, 8} 
difference = set1.difference(set2) 
print("Difference between sets:", difference)  # Output: Difference between sets: {1, 2, 3} 
```
